```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
hints =  hints.drop(['APP_REGION', 'PERSONID', 'HHID'], axis=1, inplace=False)
```

```python
#set all the features with numerical values to integers  and character as categorical

# Convert numerical features to integers
numerical_features = hints.select_dtypes(include=['int64', 'float64']).columns
hints[numerical_features] = hints[numerical_features].astype('int32')

# Convert character features to categorical
character_features = hints.select_dtypes(include=['object']).columns
hints[character_features] = hints[character_features].astype('category')
```

```python
#turn all values less than 0 to NA in integer type features
hints[numerical_features] = hints[numerical_features].applymap(lambda x: np.nan if x < 0 else x)
hints.head()
```

```python
## import here
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeClassifier

```

```python


# Data setup
# X = combined_train.drop(['EverHadCancer'], axis=1)  # Ensure combined_train is defined correctly
# y = combined_train['EverHadCancer']

X = X_resampled_train
y = Y_resampled_train
kf = KFold(n_splits=5, shuffle=True, random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# estimator1 = RandomForestClassifier(n_estimators=100, random_state=42)
# estimator2 = LinearSVC(C=1000, max_iter=100000, dual=True)
# estimator3 = Pipeline([('scale', StandardScaler()), ('knn', KNeighborsClassifier(5))])

estimator1 = DecisionTreeClassifier(random_state=42)
estimator2 = LogisticRegression(max_iter=1000, random_state=42)
estimator4 = SVC(kernel='linear', random_state=42)
estimator3 = Lasso(alpha=0.1, random_state=42)  # alpha is a tuning parameter
estimators = [estimator1, estimator2, estimator3]#, estimator4]
# estimators = [estimator3]

estimator_names = ['DecisionTree', 'LogisticRegression', 'Lasso']
# estimator_names = ['ffff']

max_features = X.shape[1]
step_size = 1

results = {}

for estimator, name in zip(estimators, estimator_names):
    average_accuracies = []
    for n_features in range(1, max_features + 1, step_size):
        selector = RFE(estimator, n_features_to_select=n_features, step=1)
        accuracies = []
        
        for train_index, test_index in skf.split(X, y):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

            # Fit RFE and transform the datasets
            X_train_rfe = selector.fit_transform(X_train, y_train)
            X_test_rfe = selector.transform(X_test)

            # Fit the model on the reduced feature set
            estimator.fit(X_train_rfe, y_train)
            predictions = estimator.predict(X_test_rfe)
            accuracy = accuracy_score(y_test, predictions)
            accuracies.append(accuracy)
        
        average_accuracy = np.mean(accuracies)
        average_accuracies.append(average_accuracy)
        print(f"{name} - Average Accuracy with {n_features} features: {average_accuracy}")

    results[name] = average_accuracies

# Plotting results
plt.figure(figsize=(10, 8))
for name, accuracies in results.items():
    plt.plot(range(1, max_features + 1), accuracies, label=name)
plt.title('Model Performance Comparison')
plt.xlabel('Number of Features Selected')
plt.ylabel('Average Accuracy')
plt.legend()
plt.show()
```
